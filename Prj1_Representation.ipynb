```python
# STEP 0 Import the necessary packages

# reading/writing image files and processing images
from skimage import io
from skimage import color

# displaying images and plots
import matplotlib.pyplot as plt

# array operations
import numpy as np
```

```python
# STEP 1 Read & display a monochrome image

# set image folder
image_folder = r'D:\yeni\Dersler\EE421-imageprocessing\images'#I changed the folder path due to error.

# choose image file
image_file = r'/cameraman.tiff'

# evaluate full path of image file
image_path = image_folder + image_file

# read color image file
imgGray = io.imread(image_path)

# display color image
plt.imshow(imgGray, cmap='gray')
plt.title('Input Color Image')
#plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# STEP 2 Extract individual bit planes of a gray scale image & display them

# extract individual bit planes
imgBP1 = imgGray.copy() & 1 # 2^0
imgBP3 = imgGray.copy() & 4 # 2^2
imgBP6 = imgGray.copy() & 32 # 2^5
imgBP8 = imgGray.copy() & 128 # 2^7

# ADD YOUR CODE HERE FOR EXTRACTING THE REMAINING BIT PLANES
imgBP2= imgGray.copy() & 2
imgBP4= imgGray.copy() & 8
imgBP5= imgGray.copy() & 16
imgBP7= imgGray.copy() & 64

# display the individual bit planes
# note that imshow by default rescales gray scale images to 255 
plt.subplot(221),plt.imshow(imgBP8, cmap = 'gray')
plt.title('Bit Plane 8'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(imgBP6, cmap = 'gray')
plt.title('Bit Plane 6'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(imgBP3, cmap = 'gray')
plt.title('Bit Plane 3'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(imgBP1, cmap = 'gray')
plt.title('Bit Plane 1'), plt.xticks([]), plt.yticks([])

# ADD YOUR CODE HERE TO DISPLAYING ALL 8 BIT PLANES IN 2x4 FORMAT
plt.subplot(241),plt.imshow(imgBP8, cmap = 'gray')
plt.title('Bit Plane 8'), plt.xticks([]), plt.yticks([])
plt.subplot(242),plt.imshow(imgBP7, cmap = 'gray')
plt.title('Bit Plane 7'), plt.xticks([]), plt.yticks([])
plt.subplot(243),plt.imshow(imgBP6, cmap = 'gray')
plt.title('Bit Plane 6'), plt.xticks([]), plt.yticks([])
plt.subplot(244),plt.imshow(imgBP5, cmap = 'gray')
plt.title('Bit Plane 5'), plt.xticks([]), plt.yticks([])
plt.subplot(245),plt.imshow(imgBP4, cmap = 'gray')
plt.title('Bit Plane 4'), plt.xticks([]), plt.yticks([])
plt.subplot(246),plt.imshow(imgBP3, cmap = 'gray')
plt.title('Bit Plane 3'), plt.xticks([]), plt.yticks([])
plt.subplot(247),plt.imshow(imgBP2, cmap = 'gray')
plt.title('Bit Plane 2'), plt.xticks([]), plt.yticks([])
plt.subplot(248),plt.imshow(imgBP1, cmap = 'gray')
plt.title('Bit Plane 1'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# STEP 3 Combine and display bit planes 5, 6, 7 and 8
img3= imgBP5+imgBP6+imgBP7+imgBP8
plt.imshow(img3,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.title('Combined bit planes 5,6,7 and 8')
plt.show()
# ADD YOUR CODE HERE
```
